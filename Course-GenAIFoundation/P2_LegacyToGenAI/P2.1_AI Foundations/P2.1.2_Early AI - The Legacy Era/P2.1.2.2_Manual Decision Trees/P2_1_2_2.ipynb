## üõ†Ô∏è  Loan Approval Decision Tree


```python
def loan_approval_decision(income, credit_score, employment_years, existing_loans, region, is_self_employed):
    # Using logical operators: and, or, not
    if (income > 80000 and credit_score > 750 and employment_years > 5 and existing_loans == 0) or (is_self_employed and income > 100000):
        return 'Approve - Premium Customer'
    elif income > 50000:
        if (credit_score > 700 and employment_years > 2) or (region in ['urban', 'metro'] and not is_self_employed):
            return 'Approve'
        elif credit_score > 650 and (region in ['urban', 'metro'] or is_self_employed) and existing_loans <= 1:
            return 'Review'
        else:
            return 'Review - Needs More Info'
    else:
        return 'Reject'

# Try different scenarios
print(loan_approval_decision(income=90000, credit_score=760, employment_years=6, existing_loans=0, region='urban', is_self_employed=False))
print(loan_approval_decision(income=120000, credit_score=680, employment_years=2, existing_loans=0, region='rural', is_self_employed=True))
print(loan_approval_decision(income=60000, credit_score=720, employment_years=3, existing_loans=1, region='metro', is_self_employed=False))
print(loan_approval_decision(income=35000, credit_score=660, employment_years=1, existing_loans=0, region='rural', is_self_employed=False))
```

## üõ†Ô∏è HR Resume Screening


```python
def resume_screening(experience_years, education_level, has_certification, has_technical_test, technical_test_score, has_portfolio):
    # Using logical operators: and, or, not
    if experience_years > 3 and (education_level in ['Masters', 'PhD'] or has_certification or has_portfolio):
        if has_technical_test and (technical_test_score >= 80 or has_portfolio):
            return 'Proceed to Interview'
        elif not has_technical_test and not has_portfolio:
            return 'Assign Technical Test'
        else:
            return 'Review - Needs More Info'
    else:
        return 'Reject'

# Try different scenarios
print(resume_screening(experience_years=5, education_level='Masters', has_certification=True, has_technical_test=True, technical_test_score=85, has_portfolio=False))
print(resume_screening(experience_years=5, education_level='PhD', has_certification=False, has_technical_test=False, technical_test_score=75, has_portfolio=False))
print(resume_screening(experience_years=2, education_level='Bachelors', has_certification=False, has_technical_test=False, technical_test_score=0, has_portfolio=False))
```

## üõ†Ô∏è Medical Diagnosis


```python
def medical_diagnosis(fever, cough, sore_throat, age, travel_history, has_rash):
    # Using logical operators: and, or, not
    if fever:
        if cough and (sore_throat or has_rash):
            if age < 12:
                return 'Pediatric Flu or Measles'
            elif travel_history and has_rash:
                return 'Possible Imported Viral Rash'
            else:
                return 'Flu'
        elif sore_throat and not cough:
            return 'Throat Infection'
        elif has_rash and not cough:
            return 'Possible Allergic Reaction'
        else:
            return 'Mild Fever - Monitor'
    else:
        return 'No major illness'

# Try different scenarios
print(medical_diagnosis(fever=True, cough=True, sore_throat=True, age=8, travel_history=False, has_rash=False))
print(medical_diagnosis(fever=True, cough=True, sore_throat=False, age=30, travel_history=True, has_rash=True))
print(medical_diagnosis(fever=True, cough=False, sore_throat=True, age=25, travel_history=False, has_rash=False))
print(medical_diagnosis(fever=True, cough=False, sore_throat=False, age=40, travel_history=False, has_rash=True))
print(medical_diagnosis(fever=False, cough=False, sore_throat=False, age=40, travel_history=False, has_rash=False))
```

## üìù Key Takeaways

- **Manual decision trees** are structured **IF‚ÄìELSE** (nested) logic ‚Äî not flat rules. Each branch can lead to more questions and outcomes.

- **Use cases:** Loan approval, HR screening, medical triage ‚Äî anywhere you have clear, layered criteria.

- **Limitations:** Still hand-crafted ‚Äî every branch and threshold is written by humans. Hard to maintain when criteria grow; no learning from data.
