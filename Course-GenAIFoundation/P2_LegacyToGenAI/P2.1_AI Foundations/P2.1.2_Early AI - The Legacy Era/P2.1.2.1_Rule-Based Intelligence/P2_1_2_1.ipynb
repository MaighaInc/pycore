## üîπ 1. IF-THEN rules ( Rule-Based System )

A rule-based system works like this:

**IF condition ‚Üí THEN action**

Example:
IF temperature > 30 ‚Üí Turn ON AC


```python
# Example: Simple thermostat rule
def thermostat(temp):
    if temp > 30:
        return "Turn ON AC"
    else:
        return "Turn OFF AC"

thermostat(35)
```

üí° This is automation using logic.


## üè• Medical Expert System

```python
def diagnose(fever, cough, body_pain, sore_throat, headache, age):
    if fever and cough and body_pain and sore_throat and headache:
        return "Severe Flu"
    elif fever and cough and sore_throat:
        return "Flu"
    elif fever and cough and headache:
        return "Viral Infection"
    elif fever and cough:
        return "Common Cold"
    elif fever and sore_throat:
        return "Throat Infection"
    elif fever and age < 12:
        return "Childhood Infection"
    elif fever:
        return "Mild Infection"
    else:
        return "No major illness"

# Try different combinations 
print(diagnose(fever=True, cough=True, body_pain=True, sore_throat=True, headache=True, age=25))
print(diagnose(fever=True, cough=True, body_pain=False, sore_throat=True, headache=False, age=30))
print(diagnose(fever=True, cough=False, body_pain=False, sore_throat=True, headache=False, age=8))
```

üè• This is how early medical AI worked.


## üí≥ Loan Approval System

```python

def loan_approval(income, credit_score, employment_years, existing_loans, region):
    if income > 80000 and credit_score > 750 and employment_years > 5 and existing_loans == 0:
        return "Loan Approved - Premium Customer"
    elif income > 50000 and credit_score > 700 and employment_years > 2:
        return "Loan Approved"
    elif income > 30000 and credit_score > 650 and region in ['urban', 'metro']:
        return "Review Required"
    elif income > 30000 and credit_score > 650:
        return "Review Required - Rural"
    else:
        return "Loan Rejected"

# Try different combinations 
print(loan_approval(income=90000, credit_score=760, employment_years=6, existing_loans=0, region='urban'))
print(loan_approval(income=60000, credit_score=720, employment_years=3, existing_loans=1, region='metro'))
print(loan_approval(income=35000, credit_score=660, employment_years=1, existing_loans=0, region='rural'))
```

‚ö†Ô∏è Notice: We are writing intelligence manually.


## üîπ Simple Chatbot

Before modern chatbots like ChatGPT, there were rule-based bots.

```python
def chatbot(user_input, context=None):
    user_input = user_input.lower()
    response = "Sorry, I don't understand."
    if "hello" in user_input or "hi" in user_input:
        response = "Hi! How can I help you?"
    elif "price" in user_input or "cost" in user_input:
        response = "Our pricing starts at $10."
    elif "bye" in user_input or "goodbye" in user_input:
        response = "Goodbye!"
    elif "help" in user_input:
        response = "You can ask about pricing, features, or support."
    elif "feature" in user_input:
        response = "We offer multiple features: A, B, C."
    elif context == "pricing":
        response = "Would you like a detailed quote?"
    elif "thanks" in user_input or "thank you" in user_input:
        response = "You're welcome!"
    return response

# Try different inputs 
print(chatbot(user_input="hello there"))
print(chatbot(user_input="Can you help me?"))
print(chatbot(user_input="What features do you offer?"))
print(chatbot(user_input="thanks!"))
print(chatbot(user_input="goodbye"))
```
