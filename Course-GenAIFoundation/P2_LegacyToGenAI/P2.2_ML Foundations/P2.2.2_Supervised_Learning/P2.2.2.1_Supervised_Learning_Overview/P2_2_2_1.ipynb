## 1Ô∏è‚É£ Supervised Learning üìò

**What is it?**

Supervised learning is a type of machine learning where the model is trained on labeled data (data with known outcomes).

**Why ‚ÄúSupervised‚Äù?**

Because we supervise the learning process by giving correct answers.

**Use:**
- Predicting outcomes for new, unseen data based on past examples.

### üîô Recap ‚Äì What we already learned

Previously we saw this example:


| X | y |
|---|---|
| 1 | 2 |
| 2 | 4 |
| 3 | 6 |
| 4 | 8 |

From this, the system learned the pattern **y = 2x**. So when we give **x = 5**, it predicts **y = 10**.

What did we actually give the machine? We gave **both**:

- **Input (X)** ‚Äî the question  
- **Correct output (y)** ‚Äî the answer  

So we gave **question + answer**. 

That‚Äôs exactly what we mean by **labeled data**, and that‚Äôs what **supervised learning** uses: examples with known answers so the model can learn, then predict for new inputs.

## üîÄ Types of problems we solve in supervised learning

In supervised learning we mainly solve **two kinds of problems**:

1Ô∏è‚É£ **Classification**  
2Ô∏è‚É£ **Regression**

üìã **Quick comparison:**

| Type | Question we answer | Output |
|------|--------------------|--------|
| **Classification** | ‚ÄúWhich category?‚Äù | A **label** (e.g. spam / not spam, cat / dog) |
| **Regression** | ‚ÄúHow much?‚Äù | A **number** (e.g. price, temperature) |

- **Classification** ‚Üí predict a **category** (discrete). Example: Is this email spam or not?
- **Regression** ‚Üí predict a **number** (continuous). Example: What‚Äôs the price of this house?


## ü§ñ Major models and algorithms

Before we close, here are some **models and algorithms** you‚Äôll often see in supervised learning. We keep it simple ‚Äî no math, no code. 

Your job is to **explore**: look up each one, see which is for **classification** vs **regression**, what **algorithm** it uses, and **why** it fits certain problems.

| Model | Usually used for | One-line idea |
|-------|------------------|----------------|
| **Linear Regression** | Regression | Fits a straight line (or hyperplane) to predict a number. |
| **Logistic Regression** | Classification | Predicts a probability, then we turn it into a category (e.g. spam / not spam). |
| **Decision Trees** | Classification & Regression | Asks yes/no questions step by step to reach a prediction. |
| **Random Forest** | Classification & Regression | Many decision trees voting together; often more robust than a single tree. |
| **k-Nearest Neighbors (k-NN)** | Classification & Regression | ‚ÄúLook at the k most similar past examples and use their answers.‚Äù |
| **Support Vector Machine (SVM)** | Classification (and some regression) | Finds a boundary that best separates classes. |
| **Naive Bayes** | Classification | Uses word (or feature) counts and simple probability; great for text (e.g. spam). |
| **Neural Networks** | Classification & Regression | Many small units connected in layers; can learn very complex patterns. |
