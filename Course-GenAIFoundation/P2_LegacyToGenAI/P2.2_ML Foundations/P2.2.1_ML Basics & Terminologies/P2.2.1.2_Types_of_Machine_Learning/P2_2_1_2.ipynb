## üß≠ The Three Types ‚Äî and What Each Needs

| Type | What you have (input) | What you need | What you get out |
|------|------------------------|---------------|------------------|
| **1Ô∏è‚É£ Supervised** | Data **with answers** (e.g. emails + spam/not spam) | Labeled examples | Predictions for new data (e.g. is this new email spam?) |
| **2Ô∏è‚É£ Unsupervised** | Data **without answers** (e.g. just customer info) | Just the data | Groups, patterns, or structure (e.g. customer segments) |
| **3Ô∏è‚É£ Reinforcement** | An **agent** in an **environment** (e.g. game, robot) | Actions + reward/penalty feedback | A policy: what action to take in each situation |

In short:
- **Supervised** = "I have examples with answers; learn and predict."
- **Unsupervised** = "I have data with no labels; find structure."
- **Reinforcement** = "I try actions and get feedback; learn what works."

## ‚≠ê Which Type is Most Commonly Used?

In practice, **Supervised Learning** is the most widely used type of machine learning for solving real-world problems. This is because most business and scientific problems have historical data with known outcomes, making supervised learning a natural fit.

- **Supervised** is the most common in practice (we often have past data with known outcomes).
