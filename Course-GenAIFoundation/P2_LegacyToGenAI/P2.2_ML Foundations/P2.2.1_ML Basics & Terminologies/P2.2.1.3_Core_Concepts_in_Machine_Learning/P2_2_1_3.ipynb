## ðŸ“‚ Train-Test Split: Why We Need It 

**What is Train-Test Split?**

- It is the process of dividing your data into two sets: one for training the model and one for testing its performance.


**Why do we need it?**

- To check if the model can generalize to new, unseen data.

**How is it done?**

- Common split: 80% training, 20% testing (can vary). 
- For example: from **100 rows** of data, you might use **80 rows** to train and **20 rows** to test.



**If we skip this step:**

- We risk building a model that works well on training data but fails on new data.
