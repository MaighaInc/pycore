## 1Ô∏è‚É£ Unsupervised Learning üìó

**What is it?**

Unsupervised learning is a type of machine learning where the model works with data that has **no labels** (no known outcomes). It finds structure or patterns in the data on its own.

**Why "Unsupervised"?**

Because we do **not** supervise with correct answers ‚Äî we only give inputs, and the model discovers groupings or patterns without being told what to look for.

**Use:**
- Discovering hidden patterns, groupings, or simpler structure in data when we don't have labels.

### üîô Recap ‚Äì How this differs from supervised learning

In **supervised learning** we had **input (X) + correct output (y)** ‚Äî question and answer. The model learned to predict y for new X.

In **unsupervised learning** we have **only input (X)** ‚Äî no answers. The model cannot "predict a label"; instead it finds **structure**: which points are similar (clustering), or how to simplify the data (dimensionality reduction). No supervision, just exploration.

## üîÄ Types of problems we solve in unsupervised learning

In unsupervised learning we mainly work with **two kinds of problems**:

1Ô∏è‚É£ **Clustering** ‚Äî group similar things together (e.g. customer segments, similar documents).  
2Ô∏è‚É£ **Dimensionality reduction** ‚Äî simplify data to fewer dimensions (e.g. for visualization, noise reduction).

üìã **Quick comparison:**

| Type | Question we answer | Output |
|------|--------------------|--------|
| **Clustering** | "Which group does this belong to?" | **Groups** (no pre-defined labels) |
| **Dimensionality reduction** | "How can we simplify this data?" | **Fewer dimensions** (e.g. for plotting or compression) |

- **Clustering** ‚Üí find **groups** in data. Example: segment customers by behavior.  
- **Dimensionality reduction** ‚Üí reduce **number of features** while keeping main structure. Example: visualize high-dimensional data in 2D.

## ü§ñ Major models and algorithms

Before we close, here are some **models and algorithms** you'll often see in unsupervised learning. We keep it simple ‚Äî no math, no code.

Your job is to **explore**: look up each one, see which is for **clustering** vs **dimensionality reduction**, and **why** it fits certain problems.

| Model | Usually used for | One-line idea |
|-------|------------------|---------------|
| **K-Means** | Clustering | Groups data into k clusters by minimizing distance to cluster centers. |
| **Hierarchical Clustering** | Clustering | Builds a tree of clusters; you can cut at different levels to get more or fewer groups. |
| **DBSCAN** | Clustering | Finds clusters based on density; can find arbitrary shapes and mark outliers. |
| **PCA** | Dimensionality reduction | Finds directions of maximum variance; projects data onto fewer dimensions. |
| **t-SNE** | Dimensionality reduction | Good for 2D/3D visualization; preserves local structure. |
