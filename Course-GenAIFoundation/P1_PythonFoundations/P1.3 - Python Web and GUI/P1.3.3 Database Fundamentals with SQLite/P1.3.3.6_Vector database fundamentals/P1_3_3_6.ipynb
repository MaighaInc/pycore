## ‚ùì What is a Vector Database?
A **vector database** stores data differently than a relational database.

Instead of tables with rows and columns, it stores **vectors** ‚Äî lists of numbers.

**Example:**
- Relational: `name = "Alice", age = 25, email = "alice@mail.com"`
- Vector: `[0.8, 0.3, 0.1, 0.9, 0.4, 0.2, ...]` (a list of numbers)
