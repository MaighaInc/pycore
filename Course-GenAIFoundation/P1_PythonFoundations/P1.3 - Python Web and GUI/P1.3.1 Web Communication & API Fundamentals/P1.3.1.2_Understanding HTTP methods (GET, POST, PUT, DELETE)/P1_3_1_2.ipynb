## üß© Before We Start: What is HTTP?
HTTP (Hypertext Transfer Protocol) is the **communication protocol** between a client and a server.
The web follows a **request ‚Üí response** model:
- A client (browser/app) sends a request
- The server returns a response
- Every action on the web is an HTTP request

## ‚ùì Why Do HTTP Methods Exist?
When a client talks to a server, it must clearly state **what action** it wants on a server.
HTTP methods are that action signal.

Examples of intent:
- ‚ÄúGet me data‚Äù
- ‚ÄúCreate something new‚Äù
- ‚ÄúUpdate existing data‚Äù
- ‚ÄúDelete a resource‚Äù

## üåê What Are HTTP Methods?
HTTP methods tell a server **what action** the client wants to perform.
Think of them as verbs in a conversation between a client (browser/app) and a server.

Common methods:
- **GET** ‚Üí read data
- **POST** ‚Üí create new data
- **PUT** ‚Üí replace/update data
- **DELETE** ‚Üí remove data

## üß† CRUD Mapping (Create, Read, Update, Delete)
| CRUD Action | HTTP Method | Example |
|---|---|---|
| Create | POST | Create a new user |
| Read | GET | Fetch user list |
| Update (replace) | PUT | Update user profile |
| Delete | DELETE | Remove a user |

## ‚úÖ Safe vs Idempotent
- **Safe**: Does *not* change server data (e.g., GET)
- **Idempotent**: Repeating the same request has the same effect (e.g., PUT, DELETE)

### Examples
- GET /users ‚Üí safe, idempotent
- POST /users ‚Üí not idempotent (creates new user each time)
- PUT /users/5 ‚Üí idempotent (replaces same resource)
- DELETE /users/5 ‚Üí idempotent (deleting twice still yields same end state)

## üîç Method-by-Method Summary
### 1) GET
- Retrieve data
- No request body (typically)
- Cachable

### 2) POST
- Create new data
- Includes a body (JSON/form data)
- Not idempotent

### 3) PUT
- Replace or update a resource
- Usually includes full object in body
- Idempotent

### 4) DELETE
- Remove a resource
- Idempotent

## ‚úÖ Key Takeaways
- HTTP methods describe **intent** (read, create, update, delete)
- GET is safe and idempotent; POST is neither
- PUT and DELETE are idempotent
- Consistent method usage makes APIs predictable and clean
