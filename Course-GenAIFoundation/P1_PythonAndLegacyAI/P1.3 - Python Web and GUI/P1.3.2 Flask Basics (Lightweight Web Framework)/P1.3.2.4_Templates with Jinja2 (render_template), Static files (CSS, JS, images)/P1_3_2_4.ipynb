## ğŸ§© What are Templates?

### Without Templates (Hard Way)
```python
@app.route('/')
def home():
    return '<html><body><h1>Hello!</h1></body></html>'
```

This is messy for large HTML pages and mixes code with markup.

### With Templates (Easy Way)
Store HTML in separate files and use Jinja2 to insert dynamic data.

### What is Jinja2?
**Jinja2** is a template engine that allows:
- Insert variables: `{{ variable_name }}`
- Use loops: `{% for item in items %}`
- Use conditionals: `{% if condition %}`
- Template inheritance: `{% extends "base.html" %}`

## ğŸ“ Basic Template Rendering

### Step 1: Create HTML Template
File: `templates/home.html`
```html
<h1>Welcome {{ username }}!</h1>
<p>Today is {{ date }}</p>
```

### Step 2: Render in Flask
```python
@app.route('/')
def home():
    return render_template('home.html', username='Alice', date='2026-02-11')
```

The `{{ }}` placeholders are replaced with actual data.

````python
from flask import Flask, render_template

app = Flask(__name__)

# Basic template with variables
@app.route('/')
def home():
    username = 'Alice'
    title = 'Welcome to My App'
    return render_template('home.html', username=username, title=title)

# Template with user profile
@app.route('/profile/<name>')
def profile(name):
    user = {
        'name': name,
        'age': 25,
        'email': f'{name}@example.com'
    }
    return render_template('user.html', user=user)

if __name__ == '__main__':
    app.run(debug=True)
````

## ğŸ”„ Loops and Conditionals in Templates

### Using For Loops
```html
<ul>
{% for item in items %}
  <li>{{ item }}</li>
{% endfor %}
</ul>
```

### Using If Conditionals
```html
{% if user.age >= 18 %}
  <p>Adult</p>
{% else %}
  <p>Minor</p>
{% endif %}
```

````python
from flask import Flask, render_template

app = Flask(__name__)

# Display list of items
@app.route('/items')
def items():
    products = ['Laptop', 'Phone', 'Tablet', 'Monitor']
    return render_template('items.html', products=products)

# Display users with conditionals
@app.route('/users')
def users():
    users_list = [
        {'name': 'Alice', 'age': 25, 'status': 'active'},
        {'name': 'Bob', 'age': 17, 'status': 'active'},
        {'name': 'Charlie', 'age': 30, 'status': 'inactive'}
    ]
    return render_template('users.html', users=users_list)

if __name__ == '__main__':
    app.run(debug=True)
````

Try:
- `http://localhost:5000/items` â†’ Shows list of products
- `http://localhost:5000/users` â†’ Shows users with age conditionals

## ğŸ¨ Static Files (CSS, JavaScript, Images)

### Linking CSS in Templates
```html
<link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">
```

### Linking JavaScript
```html
<script src="{{ url_for('static', filename='js/script.js') }}"></script>
```

### Linking Images
```html
<img src="{{ url_for('static', filename='images/logo.png') }}" alt="Logo">
```

Use `url_for('static', filename='...')` to generate correct paths.

````python
from flask import Flask, render_template, url_for

app = Flask(__name__)

# Page with CSS styling
@app.route('/')
def home():
    return render_template('styled.html')

# API endpoint for dynamic content
@app.route('/api/data')
def get_data():
    return {'message': 'Hello from API'}

if __name__ == '__main__':
    app.run(debug=True)
````

## ğŸ—ï¸ Template Inheritance

Avoid repeating HTML across pages using template inheritance.

### Base Template (base.html)
```html
<html>
<body>
  <header>My Site</header>
  {% block content %}
  {% endblock %}
  <footer>Copyright 2026</footer>
</body>
</html>
```

### Child Template (page.html)
```html
{% extends "base.html" %}
{% block content %}
  <h1>Page Title</h1>
  <p>Page content here</p>
{% endblock %}
```

````python
from flask import Flask, render_template

app = Flask(__name__)

# Home page using template inheritance
@app.route('/')
def home():
    return render_template('page_home.html', title='Home')

# About page using template inheritance
@app.route('/about')
def about():
    return render_template('page_about.html', title='About Us')

# Contact page using template inheritance
@app.route('/contact')
def contact():
    return render_template('page_contact.html', title='Contact')

if __name__ == '__main__':
    app.run(debug=True)
````

## âŒ Common Mistakes

### Mistake 1: Wrong Folder Names
âŒ Bad: Create folders named `Templates/` or `HTML/`
âœ… Good: Use `templates/` and `static/` (lowercase, exact names)

### Mistake 2: Hardcoding Paths
âŒ Bad: `<link rel="stylesheet" href="/css/style.css">`
âœ… Good: `<link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">`

### Mistake 3: Not Closing Jinja2 Tags
âŒ Bad: `{% for item in items %} ... ` (missing endfor)
âœ… Good: `{% for item in items %} ... {% endfor %}`

### Mistake 4: Using render_template Without Importing
âŒ Bad: `from flask import Flask` (missing render_template)
âœ… Good: `from flask import Flask, render_template`

### Mistake 5: Forgetting to Pass Variables
âŒ Bad: `render_template('page.html')` but template uses `{{ name }}`
âœ… Good: `render_template('page.html', name='Alice')`

## âœ… Key Takeaways
- Templates keep HTML separate from Python code
- Use `render_template()` to return HTML pages
- Insert variables with `{{ variable }}`
- Use `{% for %}` for loops and `{% if %}` for conditionals
- Use `url_for('static', filename='...')` to link static files
- Create `templates/` and `static/` folders in project root
- Use template inheritance with `{% extends %}` and `{% block %}`
- Always use lowercase folder names: `templates/` and `static/`
